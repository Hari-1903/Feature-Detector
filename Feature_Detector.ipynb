## SIFT
Scale Invariant Feature Transform (SIFT) is one of the most popular features in the field of Computer Vision.It is one of the most effective features to use for image recognition and content analysis. It is robust against scale, orientation, intensity, and so
on.

```python
import sys
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
# Load input image -- 'table.jpg'
input_file = sys.argv[1]
img = cv2.imread("/content/dog.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sift = cv2.xfeatures2d.SIFT_create()
keypoints = sift.detect(img_gray, None)
img_sift = np.copy(img)
cv2.drawKeypoints(img, keypoints, img_sift,
flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2_imshow(img)
cv2_imshow(img_sift)
cv2.waitKey()
```

## Star feature detector
Star feature Detector will give us the flexibility to cascade different blocks to get the best possible performance.

```python
import sys
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

class StarFeatureDetector(object):
    def __init__(self):
        self.detector = cv2.xfeatures2d.StarDetector_create()
    def detect(self, img):
        return self.detector.detect(img)

if __name__ == '__main__':
    input_file = sys.argv[1]
    input_img = cv2.imread("/content/dog.png") # Use input_file instead of hardcoded path
    img_gray = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
    keypoints = StarFeatureDetector().detect(img_gray) # Pass img_gray instead of input_img
    output_img = np.copy(input_img)
    cv2.drawKeypoints(input_img, keypoints, output_img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    cv2_imshow (output_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

##Histogram Equilisation
Histogram equalization is the process of modifying the intensities of the image pixels to enhance the contrast.

```python
import sys
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

input_file = "/content/dog.png"
img = cv2.imread(input_file)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)

img_gray_histeq = cv2.equalizeHist(img_gray)
cv2_imshow(img_gray_histeq)

img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
img_histeq = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)
cv2_imshow(img)
cv2_imshow(img_histeq)
cv2.waitKey()
```
